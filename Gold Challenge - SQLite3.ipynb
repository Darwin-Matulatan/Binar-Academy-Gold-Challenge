```python
# Reading file data.csv using Pandas and converting it into SQLite3

data = pd.read_csv('data.csv', encoding = 'windows-1250')

data2 = data.to_sql(name = 'Data', con = connection)
```

```python
# Displaying the whole list of Tweet column

cursorObj.execute('''select Tweet from Data''')
data3 = cursorObj.fetchall()
data3
```

```python
# Converting tuple into list in Tweet column

data4 = []

for i in data3:
    for j in i:
        data4.append(j)

data4
```

```python
# Defining a function to remove emojis and other signs

def remove_unicode_series(sentence):
    result_no_x = re.sub(r'\\x..', '', sentence)
    result_no_unicode = re.sub(r'\ufffd', '', result_no_x)
    result_no_user = re.sub(r'USER', '', result_no_unicode)
    result_no_backslash_n = re.sub(r'\\n', '', result_no_user)
    return result_no_backslash_n
```

```python
# Removing emojis and other signs from Tweet column

data5 = []

for i in data4:
    i = remove_unicode_series(i)
    data5.append(i)

data5
```

```python
# Defining a function to remove abusive words from Tweet column

def remove_abusive(sentence, abusive, replace_with = ''):
    sentence = sentence.lower()
    for i in abusive:
        sentence = re.sub(i, replace_with, sentence)
    return sentence
```

```python
# Removing abusive words from Tweet column

data6 = []

for i in data5:
    i = remove_abusive(i, abusive = abusive_spacing)
    data6.append(i)

data6
```

```python
# Defining a function to change alay words into formal words

def change_alay_to_formal(sentence, alay, formal):
    for i, j in zip(alay, formal):
        sentence = re.sub(i, j, sentence)
    return sentence
```

```python
# Converting alay words into formal words in Tweet column

data7 = []

for i in data6:
    i = change_alay_to_formal(i, alay = kamusalay_spacing, formal = kamusformal_spacing)
    data7.append(i)

data7
```

```python
# Updating contents in data7 into SQLite3 Tweet column

for i, j in zip(data6, range(len(data6))):
    connection.execute("UPDATE Data set Tweet = i where index = j")

connection.execute('''select * from Data''')
```
